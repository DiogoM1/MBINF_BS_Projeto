### 7º Iterar sobre cada gene que está no deletion_results, sempre que aparecer um gene essencial continuar o loop e depois apagar gene a gene do modelo e fazer simulação com FBA, MOMA e ROOM. Assim vamos conseguir saber a taxa de crescimento de cada mutante.

```python
from reframed import gene_knockout

for id in deletion_results.ids:
    for gene in id:
        gene = ["G_" + gene]
        if gene not in essent_genes:
            print(gene)
            print(gene_knockout(model, gene, method = 'FBA', constraints = {'R_EX_o2_e':0, 'R_EX_glc__D_e_lower_bound':-15}))
            print(gene_knockout(model, gene, method = 'MOMA', constraints = {'R_EX_o2_e':0, 'R_EX_glc__D_e_lower_bound':-15}))
            print(gene_knockout(model, gene, method = 'lMOMA', constraints = {'R_EX_o2_e':0, 'R_EX_glc__D_e_lower_bound':-15}))
            print(gene_knockout(model, gene, method = 'ROOM', constraints = {'R_EX_o2_e':0, 'R_EX_glc__D_e_lower_bound':-15}))
```
